# The Prohack Competition.

In this task we are going to predict the well being of the species. 

### Problem Statement.
optimize the well-being across all the galaxies.

### Obectives
- To understand what index makes the intergalactic species better off:
    - Use the data and train a model to predict this index with the highest possible level of certainty.

- To achieve the highest possible level of well-being through optimized allocation of additional energy:
    - A  star of an unusually high energy of 50000 zillion DSML has been discovered.
    - No one galaxy will consume more than 100 zillion DSML and,
    - at least 10% of the total energy will be consumed by galaxies in need with existence expectancy index below 0,7.

### Target variable 
y - an index of measure of well being.

### Problem Type
- Regression problem with a continuous target variable. 

### Metric
Root Mean Square Error.
